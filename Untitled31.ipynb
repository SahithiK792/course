```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import folium

# --- STEP 0: LOAD DATA ---
# This line loads the specific dataset for this assignment
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"
df = pd.read_csv(URL)

# --- 1.1: Yearly Automobile Sales Fluctuation ---
plt.figure(figsize=(10, 6))
df_line = df.groupby('Year')['Automobile_Sales'].mean().reset_index()
sns.lineplot(data=df_line, x='Year', y='Automobile_Sales')
plt.title('Automobile Sales Fluctuation from Year to Year')
plt.xlabel('Year')
plt.ylabel('Average Sales')
plt.show()


# --- 1.2: Ad Expenditure vs Sales (Non-Recession) ---
plt.figure(figsize=(10, 6))
df_non_recession = df[df['Recession'] == 0]
df_ad_sales = df_non_recession.groupby('Year')[['Automobile_Sales', 'Advertising_Expenditure']].sum().reset_index()
sns.lineplot(data=df_ad_sales, x='Year', y='Automobile_Sales', label='Sales')
sns.lineplot(data=df_ad_sales, x='Year', y='Advertising_Expenditure', label='Ad Spend')
plt.title('Advertising Expenditure vs Automobile Sales (Non-Recession)')
plt.legend()
plt.show()


# --- 1.3: Sales Trend per Vehicle Type ---
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Vehicle_Type', y='Automobile_Sales', hue='Recession')
plt.title('Sales Trend per Vehicle Type: Recession vs Non-Recession')
plt.show()


# --- 1.4: GDP Variations (Subplots) ---
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))
df_recession = df[df['Recession'] == 1]
sns.lineplot(data=df_recession, x='Year', y='GDP', ax=ax0)
ax0.set_title('GDP Variation during Recession')
sns.lineplot(data=df_non_recession, x='Year', y='GDP', ax=ax1)
ax1.set_title('GDP Variation during Non-Recession')
plt.tight_layout()
plt.show()


# --- 1.5: Seasonality Impact (Bubble Plot) ---
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_non_recession, x='Month', y='Automobile_Sales',
                size='Seasonality_Weight', hue='Seasonality_Weight', legend=False)
plt.title('Impact of Seasonality on Automobile Sales')
plt.show()


# --- 1.6: Price vs Sales Volume (Recession Only) ---
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_recession, x='Price', y='Automobile_Sales')
plt.title('Price vs Sales Volume during Recessions')
plt.show()


# --- 1.7: Ad Expenditure Portion (Pie Chart) ---
plt.figure(figsize=(8, 6))
ad_total = df.groupby('Recession')['Advertising_Expenditure'].sum()
plt.pie(ad_total, labels=['Non-Recession', 'Recession'], autopct='%1.1f%%')
plt.title('Advertising Expenditure Portion')
plt.show()


# --- 1.8: Ad Spend by Vehicle Type (Recession Only) ---
plt.figure(figsize=(8, 6))
ad_vehicle = df_recession.groupby('Vehicle_Type')['Advertising_Expenditure'].sum()
plt.pie(ad_vehicle, labels=ad_vehicle.index, autopct='%1.1f%%')
plt.title('Ad Expenditure per Vehicle Type during Recession')
plt.show()


# --- 1.9: Unemployment Rate Effect (Line Chart) ---
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_recession, x='unemployment_rate', y='Automobile_Sales', hue='Vehicle_Type')
plt.title('Unemployment Rate Effect on Sales during Recession')
plt.show()
```
